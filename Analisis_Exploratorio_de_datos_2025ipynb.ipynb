## Análisis de normalidad

### Subtask:
Para las variables cuantitativas, realizar pruebas estadísticas (como Shapiro-Wilk o Kolmogorov-Smirnov) y/o visualizaciones (histogramas, Q-Q plots) para evaluar si cumplen con el supuesto de normalidad.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Identificar variables cuantitativas (excluyendo 'Censura' y 'TJ')
quantitative_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Censura' in quantitative_cols:
    quantitative_cols.remove('Censura')
if 'TJ' in quantitative_cols:
    quantitative_cols.remove('TJ')

print("Variables cuantitativas a analizar:", quantitative_cols)

# Realizar análisis de normalidad para cada variable cuantitativa
for col in quantitative_cols:
    print(f"\nAnalizando normalidad para la columna: {col}")

    # Histograma
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

    # Q-Q plot
    plt.figure(figsize=(6, 6))
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot de {col}')
    plt.show()

    # Prueba de Shapiro-Wilk (para n <= 5000) o Kolmogorov-Smirnov (para n > 5000)
    if len(df[col]) <= 5000:
        shapiro_test = stats.shapiro(df[col])
        print(f"Prueba de Shapiro-Wilk para {col}: Estadístico={shapiro_test.statistic:.4f}, p-valor={shapiro_test.pvalue:.4f}")
        if shapiro_test.pvalue < 0.05:
            print(f"La columna '{col}' probablemente no sigue una distribución normal (p < 0.05).")
        else:
            print(f"La columna '{col}' podría seguir una distribución normal (p >= 0.05).")
    else:
        ks_test = stats.kstest(df[col], 'norm')
        print(f"Prueba de Kolmogorov-Smirnov para {col}: Estadístico={ks_test.statistic:.4f}, p-valor={ks_test.pvalue:.4f}")
        if ks_test.pvalue < 0.05:
            print(f"La columna '{col}' probablemente no sigue una distribución normal (p < 0.05).")
        else:
            print(f"La columna '{col}' podría seguir una distribución normal (p >= 0.05).")
```

## Análisis de correlación

- Para variables cualitativas y cuantitativas, evaluar la correlación utilizando métodos apropiados.

## Identificación de outliers

### Subtask:
Para cada variable cuantitativa, identificar posibles outliers utilizando métodos como el rango intercuartílico (IQR) o Z-score, y visualizarlos (box plots).

```python
# Identificar variables cuantitativas (excluyendo 'Censura' y 'TJ')
quantitative_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Censura' in quantitative_cols:
    quantitative_cols.remove('Censura')
if 'TJ' in quantitative_cols:
    quantitative_cols.remove('TJ')

print("Identificación de Outliers mediante el método IQR:")

for col in quantitative_cols:
    print(f"\nAnalizando outliers para la columna: {col}")

    # Calcular Q1, Q3 y IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Definir límites para outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identificar outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Número de posibles outliers en '{col}': {len(outliers)}")
    if len(outliers) > 0:
        print("Posibles outliers (primeras 5 filas si hay más):")
        display(outliers.head())

    # Visualizar outliers con Box Plot
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot de {col} para identificar Outliers')
    plt.xlabel(col)
    plt.show()
```
